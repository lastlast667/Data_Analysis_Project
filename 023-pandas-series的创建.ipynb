```python
'''创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。计算平均分、最高分、最低分，并找出高于平均分的学生人数。'''
np.random.seed(42)
values = np.random.randint(50,101,10)
print(values)
indexes = []
for i in range(1,11):
    indexes.append(f'学生{str(i)}')
print(indexes)
scores = pd.Series(values,indexes)
print(scores)
print('平均分：',scores.mean())
print('最高分：',scores.max())
print('最低分：',scores.min())
print(len(scores[scores > scores.mean()]))
print(scores[scores>scores.mean()].count())
```

```python
'''温度数据统计
给定某城市一周每天的最高温度Series，完成以下任务：
找出温度超过30度的天数
计算平均温度
将温度从高到低排序
找出温度变化最大的两天
'''
import pandas as pd
import numpy as np
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print('温度超过30度的天数：',temperatures[temperatures>30].count(),'天\n分别为：')
print(temperatures[temperatures>30])
print('平均温度：',temperatures.mean(),'度')
print('按温度从高到低排序',temperatures.sort_values(ascending=False))
print('==========方法一==========')
change_temp = []
for i in range(len(temperatures)):
    if i == 0:
        change_temp.append(0)
    else:
        change_temp.append(abs(temperatures.iloc[i]-temperatures.iloc[i-1]))
change_temp = pd.Series(change_temp,index=temperatures.index)
print('温度变化最大的两天：')
print(change_temp.sort_values(ascending=False).head(2))
print('==========方法二==========')
print('温度变化最大的两天：')
print(temperatures.diff().abs().sort_values(ascending=False).head(2))
```

```python
'''
股票价格分析
给定某股票连续10个交易日的收盘价Series：
计算每日收益率（当日收盘价/前日收盘价 - 1）
找出收益率最高和最低的日期
计算波动率（收益率的标准差）
'''
import pandas as pd
import numpy as np
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                   index=pd.date_range('2023-01-01', periods=10))
yield_rate = prices.pct_change()
print('每日收益率：\n',yield_rate,sep='')
print(f'收益率最高的日期:{yield_rate.idxmax().date()}，收益率：{yield_rate.max():.2%}')
print(f'收益率最低的日期:{yield_rate.idxmin().date()}，收益率：{yield_rate.min():.2%}')
print(f'波动率：{yield_rate.std():.2%}')
top3 = yield_rate.nlargest(3)
last3 = yield_rate.nsmallest(3)
print('收益率前三的日期：')
print(top3.apply(lambda x:f'{x:.2%}'))
print('收益率倒三的日期：')
print(last3.apply(lambda x:f'{x:.2%}'))
```

```python
'''销售数据分析
某产品过去12个月的销售量Series：
计算季度平均销量（每3个月为一个季度）
找出销量最高的月份
计算月环比增长率
找出连续增长超过2个月的月份
'''
import pandas as pd
import numpy as np
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                  index=pd.date_range('2022-01-01', periods=12, freq='MS'))
# 计算季度平均销量
quarterly_avg = sales.resample("QS").mean()
quarterly_avg.index = ["第一季度", "第二季度", "第三季度", "第四季度"]
print(f'季度平均销量：\n{quarterly_avg}')
# 找出销量最高的月份
max_sale = sales.max()
max_sales_month = sales.idxmax().month
print(f'销量最高的月份：{max_sales_month}月，销量是：{max_sale}')
top3 = sales.nlargest(3)
print(f'销量最高的月份：{top3.index[0].month}月，销量是：{top3.iloc[0]}')
print(f'销量前三的月份和销量：')
for idx,val in top3.items():
    print(f'{idx.month}月 销量：{val}')
# 计算月环比增长率
yield_rate = sales.pct_change()
print(f'月环比增长率：')
for i in range(len(yield_rate)):
    print(f'{yield_rate.index[i].month}月 {yield_rate.iloc[i]:.2%}')
# 找出连续增长超过2个月的月份
print('===========方法一============')
print('连续增长超过两个月的月份：')
for i in range(3,len(yield_rate)):
    if yield_rate.iloc[i-2] > 0 and yield_rate.iloc[i-1] > 0 and yield_rate.iloc[i] > 0:
        print(f'{yield_rate.index[i].month}月')
print('===========方法二============')
is_growth = yield_rate>0
continuous_growth_months = yield_rate[is_growth.rolling(3).sum()==3].index
print('连续增长超过两个月的月份：')
for month in continuous_growth_months:
    print(f'{month.month}月')
```

```python
'''每小时销售数据分析
某商店每小时销售额Series：
按天重采样计算每日总销售额
计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
找出销售额最高的3个小时
'''
import pandas as pd
import numpy as np

# 生成示例数据（24小时销售额）
hours_sales = pd.Series(
    np.random.randint(0, 100, 24),
    index=pd.date_range('2025-01-01', periods=24, freq='h')
)
# 按天重采样计算每日总销售额
day_sales = hours_sales.resample('D').sum()
print(f'每日总销售额：{day_sales.iloc[0]}')
# 计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
total_sales = day_sales.iloc[0]
business_hours_sales = hours_sales.between_time('8:00','22:00')
non_business_hours_sales = hours_sales.between_time('23:00','7:00')
print(f'营业时间销售额比例：{business_hours_sales.sum() / total_sales:.2%}')
print(f'非营业时间销售额比例：{non_business_hours_sales.sum() / total_sales:.2%}%')
# 找出销售额最高的三个小时
print('=========方法一==========')
top3 = hours_sales.nlargest(3)
print('top3:')
for time,sale in top3.items():
    print(f'{time.hour} 销量：{sale}')
print('=========方法二==========')
print(f'top3:{top3.index.hour.tolist()}')

```
