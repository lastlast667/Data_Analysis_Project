```python
'''
案例1：学生成绩分析
场景：某班级的学生成绩数据如下，请完成以下任务：
1. 计算每位学生的总分和平均分。
2. 找出数学成绩高于90分或英语成绩高于85分的学生。
3. 按总分从高到低排序，并输出前3名学生。
'''

import pandas as pd

data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '数学': [85, 92, 78, 88, 95],
    '英语': [90, 88, 85, 92, 80],
    '物理': [75, 80, 88, 85, 90]
}
scores = pd.DataFrame(data)
# 1. 计算每位学生的总分和平均分。
scores['总分'] = scores[['数学','英语','物理']].sum(axis=1)
scores['平均分'] = scores[['数学','英语','物理']].mean(axis=1).round(1)
# scores
# 2. 找出数学成绩高于90分或英语成绩高于85分的学生。
print('数学成绩高于90分或英语成绩高于85分的学生：')
condition = (scores["数学"]>90) | (scores["英语"]>85)
print(f'{scores[condition]["姓名"]}')
# 3. 按总分从高到低排序，并输出前3名学生。
top3 = scores.sort_values(by='总分', ascending=False).head(3)
print('top3：')
print(top3)
# 3. 按总分从高到低排序，并输出前3名学生
top3 = scores.nlargest(3, '总分')
print("按总分从高到低排序的前3名学生：")
print(top3)
```

```python
'''
案例2：销售数据分析
场景：某公司销售数据如下，请完成以下任务：
1. 计算每种产品的总销售额（销售额 = 单价 × 销量）。
2. 找出销售额最高的产品。
3. 按销售额从高到低排序，并输出所有产品信息。
'''

import pandas as pd

data = {
    '产品名称': ['A', 'B', 'C', 'D'],
    '单价': [100, 150, 200, 120],
    '销量': [50, 30, 20, 40]
}
df = pd.DataFrame(data)
# 1. 计算每种产品的总销售额（销售额 = 单价 × 销量）。
df['总销售额'] = df['单价']*df['销量']
print(df)
# 2. 找出销售额最高的产品。
print('销售额最高的产品：')
print(df.nlargest(1,"总销售额"))
# 3. 按销售额从高到低排序，并输出所有产品信息。
print('按销售额从高到低排序，并输出所有产品信息')
print(df.sort_values('总销售额', ascending=False))
```

```python
'''
案例3：电商用户行为分析
场景：某电商平台的用户行为数据如下，请完成以下任务：
1. 计算每位用户的总消费金额（消费金额 = 商品单价 × 购买数量）
2. 找出消费金额最高的用户，并输出其所有信息
3. 计算所有用户的平均消费金额（保留2位小数）
4. 统计电子产品的总购买数量
'''

import pandas as pd

data = {
    '用户ID': [101, 102, 103, 104, 105],
    '用户名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '商品类别': ['电子产品', '服饰', '电子产品', '家居', '服饰'],
    '商品单价': [1200, 300, 800, 150, 200],
    '购买数量': [1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)
# 1. 计算每位用户的总消费金额（消费金额 = 商品单价 × 购买数量）
df['总消费金额'] = df['商品单价']*df['购买数量']
print(df)
# 2. 找出消费金额最高的用户，并输出其所有信息
print(df.nlargest(1,'总消费金额'))
# 3. 计算所有用户的平均消费金额（保留2位小数）
avg = df['总消费金额'].mean()
print(f'所有用户的平均消费金额：{avg}')
# 4. 统计电子产品的总购买数量
values = df[df['商品类别']=='电子产品']['购买数量'].sum()
print(f'电子产品的总购买数量：{values}')
```
