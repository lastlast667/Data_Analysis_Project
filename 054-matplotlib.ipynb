```python
from cProfile import label

# 绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
from numpy.ma.core import size

rcParams['font.sans-serif'] = ['SimHei']

# 创建图表，设置大小
plt.figure(figsize=(10,5))

# 要绘图的数据
month = ['1月','2月','3月','4月']
sales = [100,200,80,130]
# 绘制折线图
plt.plot(month,sales,
         label='产品A',
         color='orange',
         linewidth=2,
         linestyle='--',
         marker='o')
# 添加标题
plt.title('2025年销售趋势',color='red',fontsize=20)
# 添加坐标轴的标签
plt.xlabel('月份',fontsize=10)
plt.ylabel('销售额（万元）',fontsize=10)
# 添加图例
plt.legend(loc='upper left',fontsize=10)
# 添加网格线
plt.grid(True,alpha=1,color='red',linestyle='--')
# plt.grid(axis='x')
# 设置刻度字体大小
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=10)
# 设置y轴的范围
plt.ylim(20,300)
# 在每个数据点上显示数值
for x,y in zip(month,sales):
    plt.text(x,y+10,str(y),fontsize=10, ha='center',va='center')
# 显示图表
plt.show()
```

```python
# 绘制柱状图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图表，设置大小
plt.figure(figsize=(10,5))

# # 要绘图的数据
subjects = ['语文','数学','英语','科学']
scores = [82,92,78,88]
# # 绘制柱状图
plt.bar(subjects,scores,
        label='小红',
        color='orange',
        linewidth=2)
# # 添加标题
plt.title('小红2025年成绩分布',color='red',fontsize=20)
# # 添加坐标轴的标签
plt.xlabel('科目',fontsize=10)
plt.ylabel('分数',fontsize=10)
# # 添加图例
plt.legend(loc='upper right',fontsize=10)
# # 添加网格线
# plt.grid(True,alpha=1,color='red',linestyle='--')
plt.grid(axis='y',alpha=0.1,color='red',linestyle='--')
# # 设置刻度字体大小
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=10)
# # 设置y轴的范围
plt.ylim(0,100)
# # 在每个数据点上显示数值
for x,y in zip(subjects,scores):
    plt.text(x,y+3,str(y),fontsize=10, ha='center',va='center')
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 绘制饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图表，设置大小
plt.figure(figsize=(10,5))

# # 要绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
colors = ['red','green','blue','yellow','orange']
# # # 绘制饼图
plt.pie(times,labels=things,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors
        )
# # # 添加标题
plt.title('一天的时间分布',color='red',fontsize=20)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 绘制环形图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图表，设置大小
plt.figure(figsize=(10,5))

# # 要绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
colors = ['red','green','blue','yellow','orange']
# # # 绘制环形图
plt.pie(times,labels=things,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops={'edgecolor':'red', # 设置圆环参数
                    'width':0.5
                    },
        pctdistance=0.8,
        )
# # # 添加标题
plt.title('一天的时间分布',color='red',fontsize=20)
plt.text(0,0,'总计：\n100%',ha='center',va='center')
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 绘制爆炸式饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图表，设置大小
plt.figure(figsize=(10,5))

# # 要绘图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
colors = ['red','green','blue','yellow','orange']
explode = [0.2,0,0,0,0]
# # # 绘制爆炸式饼图
plt.pie(times,labels=things,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        pctdistance=0.8,
        explode=explode,
        shadow=True,
        )
# # # 添加标题
plt.title('一天的时间分布',color='red',fontsize=20)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 绘制散点图
import matplotlib.pyplot as plt
from matplotlib import rcParams # 字体
import random
rcParams['font.sans-serif'] = ['SimHei']
rcParams['axes.unicode_minus'] = False
# 创建图表，设置大小
plt.figure(figsize=(10,8))

# # 要绘图的数据
x=[]
y=[]
for i in range(1000):
    tmp = random.uniform(0,10)
    x.append(tmp)
    tmp2 = 2*tmp + random.gauss(0,3)
    y.append(tmp2)

# # # 绘制散点图
plt.scatter(x,y,
            color='blue',
            alpha=0.8,
            s=20,
            label='数据')
# # 添加标题
plt.title('x变量与y变量的关系',color='red',fontsize=20)
# # 添加坐标轴的标签
plt.xlabel('X自变量',fontsize=10)
plt.ylabel('Y因变量',fontsize=10)
# # # 添加图例
plt.legend(loc='upper right',fontsize=10)
# # # 添加网格线
plt.grid(True,alpha=0.5,color='red',linestyle='--')
# plt.grid(axis='y',alpha=0.1,color='red',linestyle='--')
# # # 设置刻度字体大小
plt.xticks(fontsize=20,rotation=0)
plt.yticks(fontsize=10)
# # # 设置y轴的范围
plt.ylim(0,30)
plt.plot([0,10],[0,20],color='red',linewidth=2)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

```python
# 分析案例：温度分析
# 1.导入库
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['SimHei']
# 导入数据
df = pd.read_csv('data/weather.csv')
df.head()

# 绘制气温的趋势变化图
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year==2015]
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_max'],label='最高气温')
plt.plot(df['date'],df['temp_min'],label='最低气温')
plt.title('2015年气温趋势变化图',color='red',fontsize=30)
plt.xlabel('日期',fontsize=20)
plt.ylabel('气温',fontsize=15)
plt.legend(fontsize=20)
```

```python
# 绘制平均气温的趋势变化图
df['temp_mean'] = (df['temp_max'] + df['temp_min']) / 2
df = df[df['date'].dt.year==2015]
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_mean'],label='平均气温')
plt.title('2015年平均气温趋势变化图',color='red',fontsize=30)
plt.xlabel('日期',fontsize=20)
plt.ylabel('气温',fontsize=15)
plt.legend(fontsize=20)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
penguins = pd.read_csv('data/penguins.csv')
penguins.dropna(inplace=True)
penguins.info()
penguins.head()
```

```python
# 直方图
sns.histplot(data=penguins,x='species',hue='species',palette=sns.color_palette())
```

```python
# 核密度估计图
# 绘制喙长度的核密度估计图
sns.kdeplot(data=penguins,x='bill_length_mm')
```

```python
sns.histplot(data=penguins,x='bill_length_mm',hue='species',kde=True)
```

```python
# 计数图
# 绘制不同岛屿企鹅数量的计数图
sns.countplot(data=penguins,x='island')
```

```python
sns.histplot(data=penguins,x='island')
```

```python
# 散点图
sns.scatterplot(data=penguins,x='body_mass_g',y='flipper_length_mm',hue='sex')
```

```python
# 蜂窝图
sns.jointplot(data=penguins,x='body_mass_g',y='flipper_length_mm',kind='hex')
```

```python
# 成对关系图
sns.pairplot(data=penguins,hue='species')
```

```python
# 小提琴图
sns.violinplot(data=penguins,x='species',y='bill_length_mm')
```
